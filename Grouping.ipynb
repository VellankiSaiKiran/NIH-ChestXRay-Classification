```python
import os
import shutil

# Assuming 'df_filtered' is the DataFrame from the previous step and it's already in memory

source_folder = '/Users/saikiranreddyvellanki/Documents/CapstoneProject/resized_images'  # Your source folder containing images
destination_root = '/Users/saikiranreddyvellanki/Documents/CapstoneProject/Images'  # The root folder where you want to create the disease folders

# Make sure the destination root exists
os.makedirs(destination_root, exist_ok=True)

# Iterate through the DataFrame rows
for index, row in df_filtered.iterrows():
    file_name = row['id']  # Get the file name
    disease = 'No_Disease'  # Default folder for images with no disease
    
    # If there is a disease present, get the disease name
    diseases_in_row = row[1:]  # Adjust the slice if 'id' is not the first column
    # Check for the presence of a disease and get its name
    for disease_name in diseases_in_row.index:
        if row[disease_name] == 1:
            disease = disease_name
            break  # If a disease is found, no need to check further

    # Define source and destination paths
    source_path = os.path.join(source_folder, file_name)
    disease_folder = os.path.join(destination_root, disease)
    
    # Create a folder for the disease if it does not exist
    os.makedirs(disease_folder, exist_ok=True)
    
    destination_path = os.path.join(disease_folder, file_name)
    
    # Move the file from source to destination
    shutil.move(source_path, destination_path)
```

```python
import os
from pathlib import Path

# Path to the directory where the class-specific folders are saved
output_directory = Path('/Users/saikiranreddyvellanki/Documents/CapstoneProject/Images')

# Dictionary to hold the count of images in each class directory
class_image_counts = {}

# Iterate over each class directory in the output directory
for class_directory in output_directory.iterdir():
    if class_directory.is_dir():  # Check if it is a directory
        # Count the number of files in the directory
        class_image_count = len(list(class_directory.glob('*.*')))
        # Add the count to the dictionary
        class_image_counts[class_directory.name] = class_image_count

# Print the counts for each class
for class_name, count in class_image_counts.items():
    print(f'{class_name}: {count} images')
```

```python
import os

root_directory = '/Users/saikiranreddyvellanki/Documents/CapstoneProject/Images'  # Replace with the path to your folders

# Supported image file extensions
image_extensions = {'.jpg', '.jpeg', '.png', '.gif'}

# Initialize a counter
total_images = 0

# Walk through all directories and files in the root directory
for root, dirs, files in os.walk(root_directory):
    # Count only files with supported image extensions
    total_images += sum(file.endswith(ext) for file in files for ext in image_extensions)

print(f"Total number of images in all folders: {total_images}")
```
