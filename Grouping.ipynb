```python
import os
import pandas as pd
import zipfile
from pathlib import Path
from shutil import copyfileobj



# Load the labels DataFrame
df = concatenated_labels_df_loaded

# Directory where you want to save the class-specific folders
output_directory = Path('/Users/saikiranreddyvellanki/Documents/CapstoneProject/Files')
output_directory.mkdir(parents=True, exist_ok=True)

# Process the images from the ZIP file
with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    # Iterate over each class (column) in the DataFrame except the 'id' column
    for class_name in df.columns[1:]:
        # Create a directory for this class
        class_directory = output_directory / class_name
        class_directory.mkdir(exist_ok=True)
        
        # Get the image filenames where the class is present
        image_filenames = df[df[class_name] == 1]['id']
        
        # Extract and save these images into the class-specific directory
        for filename in image_filenames:
            image_path_in_zip = f'resized_images/{filename}'  # Adjust this if the path in the zip is different
            try:
                # Open the image file from the zip file
                with zip_ref.open(image_path_in_zip) as image_file:
                    # Create a new file path in the class directory
                    destination_path = class_directory / filename
                    # Copy the image file to the new directory
                    with open(destination_path, 'wb') as new_file:
                        copyfileobj(image_file, new_file)
            except KeyError:
                print(f"Image {filename} not found in the zip file.")
```

```python
import os
from pathlib import Path

# Path to the directory where the class-specific folders are saved
output_directory = Path('/Users/saikiranreddyvellanki/Documents/CapstoneProject/Files')

# Dictionary to hold the count of images in each class directory
class_image_counts = {}

# Iterate over each class directory in the output directory
for class_directory in output_directory.iterdir():
    if class_directory.is_dir():  # Check if it is a directory
        # Count the number of files in the directory
        class_image_count = len(list(class_directory.glob('*.*')))
        # Add the count to the dictionary
        class_image_counts[class_directory.name] = class_image_count

# Print the counts for each class
for class_name, count in class_image_counts.items():
    print(f'{class_name}: {count} images')
```
